### Custom Resampling Functions

```python
#We can define our own function
def first_day(entry):
    """
    Returns the first instance of the period, regardless of sampling rate.
    """
    
    if len(entry):  # handles the case of missing data
        return entry[0]
```

```python
df.resample(rule='A').apply(first_day)
```

Y_t=T_t + C_t     (Equation)

It essentially seperates out the data into 2 components,it tries to find the trend component and cyclical component of this data source and those components are determined by minimizing the following quadratic loss function where lambda is a smoothing parameter.

SO this is the essentially equation that

## Using Statsmodels to get the trend
<div class="alert alert-info"><h3>Related Function:</h3>
<tt><a href='https://www.statsmodels.org/stable/generated/statsmodels.tsa.filters.hp_filter.hpfilter.html'><strong>statsmodels.tsa.filters.hp_filter.hpfilter</strong></a><font color=black>(X, lamb=1600)</font>&nbsp;&nbsp;Hodrick-Prescott filter</div>
    
The <a href='https://en.wikipedia.org/wiki/Hodrick%E2%80%93Prescott_filter'>Hodrick-Prescott filter</a> separates a time-series  $y_t$ into a trend component $\tau_t$ and a cyclical component $c_t$

$y_t = \tau_t + c_t$

The components are determined by minimizing the following quadratic loss function, where $\lambda$ is a smoothing parameter:

$\min_{\\{ \tau_{t}\\} }\sum_{t=1}^{T}c_{t}^{2}+\lambda\sum_{t=1}^{T}\left[\left(\tau_{t}-\tau_{t-1}\right)-\left(\tau_{t-1}-\tau_{t-2}\right)\right]^{2}$


The $\lambda$ value above handles variations in the growth rate of the trend component.<br>When analyzing quarterly data, the default lambda value of 1600 is recommended. Use 6.25 for annual data, and 129,600 for monthly data.

```python
from statsmodels.tsa.filters.hp_filter import hpfilter
gdp_cycle,gdp_trend=hpfilter(mdf['realgdp'],lamb=1600)
```

```python
gdp_trend.plot()
```

```python
mdf['trend']=gdp_trend  # gdp_trend represents pandas series
```

```python
mdf[['trend','realgdp']]['2000-03-31':].plot(figsize=(12,8)).autoscale(axis='x',tight=True);
```

# ETS Decomposition model(Error-Trend-Seasonality)

1.ETS is able to pull out the trend, seasonality and residual/error term<br>
2.trend term is showing you the general growth pattern or decline pattern depending on the dataset of the actual real observed data points.To generally check whether pattern is linear or exponential.<br>
3.Removing the trend we will get the seasonality!<br>
4.last residual component is for things not explained by the trend or seasonality.Anything that is not explained by trend and seasonality.From residual graph we get to know whether it is containing noises or not



Two types of main ETS model<br>
1.Additive Model -> when it seems that the trend is more linear and the seasonality & Trend components seem to be constant over time. For ex- every year we add 10000 passengers<br>
2.Multiplicative model is more appropriate when we are increasing (or decreasing ) at a non linear rate(e.g. each year we are doubling the amount of passengers).

# EWMA Models (Exponentially weighted moving average models)

Basic SMA(simple moving average) has some weakness ---<br>
1.smaller window will lead to more noise rather than a signal.<br>
2.it will always lag by the size of window<br>
3.it will never reach to full peak or valley of the seasonal data due to the averaging even if we select the small window<br>
4.Does not really inform you about possible future behaviour,all it really does is describe trends in your data.<br>
5.Extreme historical values can skew your SMA significantly.(like recession)<br>

**To help fix all this issues we can use EWMA(Exponentially weighted moving average)**<br>

EWMA will allows us to reduce the lag effect from SMA and it will put more weight on values  that occured more recently(by applying more weight to the recent values ).

The amount of weight applied to the recent values will depend on the actual parameters used in EWMA and the number of periods given a window size.
